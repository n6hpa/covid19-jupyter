```python
import scipy as sp
import numpy as np
import pandas as pd
import os
from datetime import datetime

from bokeh.io import push_notebook, show, output_notebook
from bokeh.layouts import row
from bokeh.plotting import figure
output_notebook()
```

```python
# Setup          /home/jovyan/work/covid19-jupyter/covid19-data/data/csv
base_data_dir = 'covid19-data/data/csv/'
dates = pd.date_range('2020-03-14', periods=6).to_list()

date_strs = [str(date.date()) for date in dates]
print(date_strs)
paths = [os.path.join(base_data_dir, date+'.csv') for date in date_strs]

data_frames = [pd.read_csv(path) for path in paths]

state_name = 'California'

state_data = [data_frame[data_frame['State_Name']==state_name] for data_frame in data_frames]

total_confirmed = state_data[-1]['Confirmed'].sum()

state_data[-1].head()

county_names = state_data[-1]['County_Name'].tolist()

total_cases_time_series_frame = pd.DataFrame({'Date': dates})

frame = state_data[-1]

for county_name in county_names:
    data = []
    for frame in state_data:
        data_frame = frame[frame['County_Name']==county_name]
        if data_frame.empty:
            data.append(0)
        else:
            data.append(int(data_frame['Confirmed']))
    total_cases_time_series_frame[county_name] = data
    
total_cases_time_series_frame.tail()
```
