```python
import scipy as sp
import numpy as np
import pandas as pd
import os
from datetime import datetime

import covid19_jupyter

from bokeh.io import push_notebook, show, output_notebook
from bokeh.layouts import row
from bokeh.plotting import figure
output_notebook()
```

```python
combined_data = covid19_jupyter.load_combined_data('covid19-data/data/csv')
combined_data.tail()
# State Data Test
state_data = covid19_jupyter.get_state_data(combined_data, 'California')

total_cases_time_series_frame = covid19_jupyter.tranform_state_to_time_series(state_data)
total_cases_time_series_frame.tail(10)
```

```python
p1 = figure(plot_width=800, plot_height=500, x_axis_type='datetime')
p1.circle(total_cases_time_series_frame.index, total_cases_time_series_frame['San Francisco'])
p1.circle(total_cases_time_series_frame.index, total_cases_time_series_frame['Alameda'])
show(p1)
```
